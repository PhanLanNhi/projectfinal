```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Đọc dữ liệu đã xử lý
df = pd.read_csv(r'D:\dow\project_final\data\processed_iot_dataset_final.csv')

# (Tuỳ chọn: chỉ giữ những Sensor_ID hợp lệ)
sensor_id_col = 'Sensor_ID'
if sensor_id_col in df.columns:
    valid_sensor_ids = df[(df[sensor_id_col] >= 100) & (df[sensor_id_col] <= 200)][sensor_id_col].unique()
    df = df[df[sensor_id_col].isin(valid_sensor_ids)].reset_index(drop=True)

# Chia train-test
train_df, test_df = train_test_split(
    df, test_size=0.2, stratify=df['Fault_Type'], random_state=42
)

# Xuất file csv
train_df.to_csv(r'D:\dow\project_final\data\iot_train.csv', index=False)
test_df.to_csv(r'D:\dow\project_final\data\iot_test.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu
df = pd.read_csv(r'D:\dow\project_final\data\iot_train.csv')

# Vẽ biểu đồ phân phối nhãn Fault_Status
plt.figure(figsize=(7,4))
df['Fault_Status'].value_counts().sort_index().plot(kind='bar')
plt.title('Phân phối số lượng Fault_Status')
plt.xlabel('Fault_Status')
plt.ylabel('Số lượng mẫu')
plt.tight_layout()
plt.show()
```
